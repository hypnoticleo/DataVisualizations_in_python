### Scatter Graphs

```python
x1 = np.array([250,150,350,252,450,550,455,358,158,355])

y1 = np.array([40,50,80, 90, 100,50,60,88,54,45])

x2 = np.array([200,100,300,220,400,500,450,380,180,350])

y2 = np.array([400,500,800, 900, 1000,500,600,808,504,405])

# Graph 1
plt.scatter(x1,y1, color = '#fc1505')
plt.title('Sactter graph')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()

# Graph 2
plt.scatter(x2,y2, color = '#057cfc')
plt.title('Sactter graph')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()

# Graph 3
plt.scatter(x2,y2, color = '#057cfc')
plt.scatter(x1,y1, color = '#fc1505')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.show()

# Graph 4
plt.scatter(x1,y1, label = 'Parameter 1', marker='^', color='#fc2e05')
plt.scatter(x2, y2, label = 'Parameter 2', marker='*', color='#fc4b05')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.legend()
plt.show()
```
