## Data Visualizations With Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-darkgrid')
```

## Area Plots

```python
plt.figure(figsize=(16,6))
x = np.arange(0, 30, 1)
y = np.random.normal(10, 11,30)
plt.plot(x,y)
plt.fill_between(x,y)
plt.show()
```

## Custom Colors 

```python
plt.figure(figsize=(16,6))
x = np.arange(0, 30, 1)
y = np.random.normal(10, 11,30)
plt.plot(x,y)
plt.fill_between(x,y, color = '#24d8cf')
plt.title(" Area Chart" , fontsize = 16)
plt.xlabel("X-Axis" , fontsize = 16)
plt.ylabel("Y-Axis" , fontsize = 16)
plt.show()
```

## Transperancy Settings

```python
plt.figure(figsize=(16,6))
x = np.arange(0, 30, 1)
y = np.random.normal(10, 11,30)
plt.plot(x,y)
plt.fill_between(x,y, color = '#24d8cf', alpha = 0.2)
plt.title(" Area Chart" , fontsize = 16)
plt.xlabel("X-Axis" , fontsize = 16)
plt.ylabel("Y-Axis" , fontsize = 16)
plt.show()

plt.figure(figsize=(16,6))
x = np.arange(0, 30, 1)
y = np.random.normal(10, 11,30)
plt.plot(x,y)
plt.fill_between(x,y, color = '#24d8cf', alpha = 0.6)
plt.title(" Area Chart" , fontsize = 16)
plt.xlabel("X-Axis" , fontsize = 16)
plt.ylabel("Y-Axis" , fontsize = 16)
plt.show()
```

## Stacked Area Plot

```python
x = np.arange(1,7)
a = np.array([1,5,9,13,17,21]) 
b = np.array([2,6,10,14,18,22])
c = np.array([3,7,11,15,19,23])
d = np.array([4,8,12,16,20,24])

plt.figure(figsize=(10,8), dpi=100)
plt.stackplot(x,a,b,c,d, labels=['A','B','C','D'])
plt.legend(loc = 'upper left')
plt.show()
```

```python
x = np.arange(1,7)

y = [[1,5,9,13,17,21],[2,6,10,14,18,22],[3,7,11,15,19,23],[4,8,12,16,20,24]]
plt.figure(figsize=(10,8), dpi=100)
plt.stackplot(x,y, labels=['A','B','C','D'])
plt.legend(loc = 'upper left')
plt.show()
```

```python
x=np.arange(1,7)
y=[ [1,5,9,3,17,1], [2,6,10,4,16,2], [3,7,11,5,19,1] , [4,8,12,6,20,2] ]

plt.figure(figsize=(11,6))
plt.stackplot(x,y , labels=['Y1','Y2','Y3','Y4'] , colors= ["#00b159" , "#ffc425", "#f37735", "#ff3b30"])
plt.legend(loc='upper left')
plt.show()

plt.figure(figsize=(11,6))
plt.stackplot(x,y, labels=['Y1','Y2','Y3','Y4'], colors= ["#00b159" , "#ffc425", "#f37735", "#ff3b30"], alpha=0.7 )
plt.legend(loc='upper left')
plt.show()

plt.figure(figsize=(11,6))
plt.stackplot(x,y, labels=['Y1','Y2','Y3','Y4'], colors= ["#00b159" , "#ffc425", "#f37735", "#ff3b30"], alpha=0.5 )
plt.legend(loc='upper left')
plt.show()
```
