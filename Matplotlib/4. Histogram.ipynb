```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use('seaborn-darkgrid')
```

### Histograms

```python
plt.figure(figsize=(8,5), dpi=100)


x =  np.random.normal(size = 2500)
plt.hist(x, bins=28, color='#45e519')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
plt.figure(figsize=(8,5), dpi=100)


x =  np.random.uniform(size = 2500)
plt.hist(x, bins=40, color='#c90835')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
plt.figure(figsize=(8,5), dpi=100)


x =  np.random.normal(size = 2500)
plt.hist(x, bins=40, color='#fc0c10', edgecolor = '#c63b3d')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

### Multiple Bin Size

```python
x =  np.random.normal(size = 2500)

plt.figure(figsize=(10,8), dpi=100)
plt.hist(x, bins=100, color='#f7860e', edgecolor = '#3f3b37')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')

plt.figure(figsize=(10,8), dpi=100)
plt.hist(x, bins=50, color='#f7860e', edgecolor = '#3f3b37')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')

plt.figure(figsize=(10,8), dpi=100)
plt.hist(x, bins=20, color='#f7860e', edgecolor = '#3f3b37')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')

plt.show()
```

## Multiple Plots In One Graph

```python
plt.figure(figsize=(10,8), dpi = 100)
a = np.random.normal(loc = -5, scale = 1 , size=1000)
b = np.random.normal(loc = 0, scale = 1 , size=1000)
c = np.random.normal(loc = 5, scale = 1 , size=1000)
plt.hist(a, bins=40, color='#fc0c10', edgecolor = '#c63b3d')
plt.hist(b, bins=40, color='#8bf22b', edgecolor = '#5a965b')
plt.hist(c, bins=40, color='#1103ad', edgecolor = '#46427a')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')

plt.show()
```

### Cumulative Histogram

```python
plt.figure(figsize=(10,8), dpi=100)
x =  np.random.normal(size = 1000)
plt.hist(x, bins=50, color='#f7860e', edgecolor = '#3f3b37', cumulative = 1)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')
```
