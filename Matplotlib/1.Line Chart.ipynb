### Line Graphs

 by default the plot function will plot a line graph

```python
x = np.arange(0,15,1)
y = np.power(x,2)
plt.plot(x,y)
plt.show()
```

### Custom aspect ratio for the graph

```python
plt.figure(figsize = (8,5), dpi=100)
x = np.linspace(0,10,1000)
y = np.cos(x+90)
plt.plot(x,y)
plt.show()
```

### Customizing the color of graph

```python
plt.figure(figsize = (8,5), dpi=100)
x = np.linspace(0,10,1000)
y = np.cos(x+90)
plt.plot(x,y, 'r-')
plt.show()
```

```python
plt.figure(figsize = (8,5), dpi=100)
x = np.linspace(0,10,1000)
y = np.cos(x+90)
plt.plot(x,y, color = '#d60a3a')
plt.title('Cosine graph ')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
```

### Plotting multiple Lines 

```python
plt.figure(figsize = (8,5), dpi=100)
x = np.arange(0,20,1)
y1 = np.power(x, 2)
y2 = np.power(x, 2.1)
plt.plot(x,y1, label = 'x^2')
plt.plot(x,y2, label = 'x^2.1')
plt.title('Growth graph ')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.show()
```

```python
x = np.linspace(0,10,1000)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y1, label = 'Sin(x)', color = '#0a7ad6')
plt.plot(x,y2, label = 'Cos(x)', color = '#5fd60a')
plt.title('Sin(x) & Cos(x)')
plt.legend(bbox_to_anchor = (1.2,1))
plt.show()
```

### line styles

```python
plt.figure(figsize=(16,9), dpi= 100)
x = np.linspace(0,20,100)
y1 = np.sin(x)
y2 = np.sin(x+180/2)
y3 = np.sin(x+180/3)
y4 = np.sin(x+180/4)


plt.plot(x,y1, label = 'dashed' , color = 'red', linestyle = '--', linewidth = 1)
plt.plot(x,y2, label = 'dashdot' , color = 'green', linestyle = '-.', linewidth = 1)
plt.plot(x,y3, label = 'solid' , color = 'blue', linestyle = '-', linewidth = 1)
plt.plot(x,y4, label = 'dotted', color = 'magenta', linestyle = ':',linewidth = 1)

plt.legend(bbox_to_anchor = (1,1))
plt.show()
```

### Shading regions

```python
x = np.linspace(0,10,1000)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y1, label = 'Sin(x)', color = '#0a11d6')
plt.plot(x,y2, label = 'Cos(x)', color = '#ff3a28')
plt.title('Sin(x) & Cos(x)')
plt.fill_between(x,y1)
plt.fill_between(x,y2)
plt.legend(bbox_to_anchor = (1.2,1))
plt.show()
```

### Moving the Legend Position

```python
x = np.linspace(0,10,1000)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y1, label = 'Sin(x)', color = '#0a7ad6')
plt.plot(x,y2, label = 'Cos(x)', color = '#5fd60a')
plt.title('Sin(x) & Cos(x)')
plt.legend(loc = 'upper right', fontsize = 'small')
plt.show()
```

### Plotting with Grid Lines

```python
x = np.arange(0,20,1)
y = 2*x+3
z = 4*x+5
plt.figure(figsize = (10,5))
plt.plot(x, y, color = '#ff076e')
plt.plot(x, z)

plt.grid(b = True,linestyle = '-', which='both', color = '#000000',linewidth = 0.2)

plt.show()
```

Setting the background color

```python
x = np.arange(0,20,1)
y = 2*x+3
z = 4*x+5

plt.figure(figsize = (10,5))
ax = plt.axes()
ax.set_facecolor('#eae8e9')

plt.plot(x, y)
plt.plot(x, z)

plt.show()
```

### Subplot Example

```python
x = np.arange(0,20,1)
y = 2*x
y2 = x**4

plt.figure(figsize=(8,3))
plt.subplot(1,2,1)
plt.plot(x,y)

plt.subplot(1,2,2)
plt.plot(x,y2)

plt.show()
```

```python
x = np.linspace(0, 10, 1000)
y = np.sin(x)
y1 = np.cos(x)


plt.figure(figsize = (16,9))

plt.subplot(2,1,1)
plt.plot(x,y)

plt.subplot(2,1,2)
plt.plot(x,y1)

plt.show()
```
